```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([30, 50, 70, 80, 100])
x2 = np.array([0,1,1,2,1])
y = np.array([0, 1, 0, 1, 1])
Idonate = np.where(y==1)[0]
Inodonate = np.where(y==0)[0]

plt.plot(x1[Idonate],x2[Idonate],'g.')
plt.plot(x1[Inodonate],x2[Inodonate],'r.')
plt.xlabel("Income")
plt.ylabel("Number of Websites")
plt.xlim(25,125)
plt.legend(['donated','no donated'],loc='upper right')
plt.show()
```
