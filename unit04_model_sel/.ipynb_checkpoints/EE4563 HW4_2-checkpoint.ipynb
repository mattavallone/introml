```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0,1,10) # some uniform x data
y0 = poly.polyval(x,beta_0)
ydat = y0 + np.random.normal(0,0,10)
plt.scatter(x,ydat)
plt.show()

x_tr = np.linspace(0,1,10)
d = 2 # degree of the polynomial
beta_hat = poly.polyfit(x_tr,ydat,d)
y_hat = poly.polyval(x_tr,beta_hat) #calculating y estimate
plt.plot(x_tr,y_hat)
plt.show()
```

```python
y = poly.polyval(x_tr,beta_0)
plt.xlim(0,3)
plt.plot(x_tr,y,'r-') #plotting true function
plt.plot(x_tr,y_hat,'g-') #plotting estimated function
plt.grid()
plt.legend(['True (dtrue=2)', 'Est (d=2)'], loc='upper left')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
